# Ridge Regression

```python
from si.io.CSV import read_csv
from si.linear_model.ridge_regression import RidgeRegression
from si.linear_model.logistic_regression import LogisticRegression
from si.model_selection.split import train_test_split
from sklearn import preprocessing

import matplotlib.pyplot as plt
```

```python
data = read_csv("C:/Users/rober/si/datasets/cpu/cpu.csv", ",", True, -1)

data_train, data_test = train_test_split(data, 0.3, 1)
```

### Regular fit()

```python
model = RidgeRegression(max_iter=2000)
model.fit(data_train, use_adaptive_fit=False)
print("Predicted test values:\n----------------------")
print(model.predict(data_test))
print("\nScore:\n------")
print(model.score(data_test))
print("\nCost:\n-----")
print(model.cost(data_test))
```

```python
plt.plot(list(model.cost_history.keys()), list(model.cost_history.values()), '-', color="yellow")
plt.title("Cost History")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show
```

### Adaptive fit()

```python
model = RidgeRegression(max_iter=2000)
model.fit(data_train, use_adaptive_fit=True)
print("Predicted test values:\n----------------------")
print(model.predict(data_test))
print("\nScore:\n------")
print(model.score(data_test))
print("\nCost:\n-----")
print(model.cost(data_test))
```

```python
plt.plot(list(model.cost_history.keys()), list(model.cost_history.values()), '-', color="yellow")
plt.title("Cost History")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show
```
